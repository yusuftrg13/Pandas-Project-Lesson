```python
import pandas as pd
import numpy as np

#Bu veriler gerçek bir veri seti olmamakla birlikte herhangi bir tavsiye değildir.Sadece tekrar maksatlı yaptığım bir projedir
# Although these data are not a real data set, they are not any recommendations. It is just a project I made for repeat purposes.
```

```python
np.random.seed(0)
```

```python
#Yıllar
years=range(2000,2022)
```

```python
#Kurgusal veri seti oluşturalım

food_prices = {
    'Year':years,
    'Wheat':np.random.normal(200,10,len(years)),
    'Rice':np.random.normal(150,8,len(years)),
    'Corn':np.random.normal(180,12,len(years)),
    'Soybean':np.random.normal(250,15,len(years))
}

#DataFrame oluşturma
df=pd.DataFrame(food_prices)
print(df.head())
```

```python
#Bilgileri gösterelim
print(df.info())
```

```python
#Temel istatistik
print(df.describe())
```

```python
#2012 yılı veri bilgileri
year_2012=df[df['Year']==2012]
print(year_2012)
```

```python
#Belirli bir ürüne ait fiyatların ortalama değerini alalım

avg_wheat_price=df['Wheat'].mean()
print("Ortalama Buğday Fiyatı:",avg_wheat_price)
```

```python
df.plot(x='Year', y='Wheat', kind='line', title='Buğday Fiyatları')
```

```python
import matplotlib.pyplot as plt

# Matplotlib ile çizgi grafiği oluşturma
plt.figure(figsize=(10, 6))  # Şekil boyutunu belirleme
plt.plot(df['Year'], df['Wheat'], label='Wheat')
plt.plot(df['Year'], df['Rice'], label='Rice')
plt.plot(df['Year'], df['Corn'], label='Corn')
plt.plot(df['Year'], df['Soybean'], label='Soybean')
plt.title('Global Food Prices Over Years')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

```
